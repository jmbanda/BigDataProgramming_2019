```
data = spark.read.csv('USDA_activity_dataset.csv',inferSchema=True, header=True)
```

Let's explore the data file.

```
data.count(), len(data.columns)
```

Let's see the data

```
data.show(5)
```

Let's see the data schema.

```
data.printSchema()
```

Now time for some column filtering.

```
data.select("State","County","Median household income","Poverty rate") \
.show(15, truncate=False)
```
